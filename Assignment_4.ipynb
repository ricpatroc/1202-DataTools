The original Data fram has some **NaN** values. we need to drop these rows before selecting the first 1000 channels

```python
###Dropping NaN values only by Channel type
channel_frame = data_original.dropna(subset=['channeltype'])
channel_frame.head(20)
```

```python
#Select the first 1000(0-999) channels based on subscribers and count the specific values of the channeltype column
channels = channel_frame.channeltype[:999].value_counts()
```

```python
dataframe  = pd.DataFrame(channels)
#channels.count()
```

```python
dataframe.plot(kind='pie', y = 'channeltype', autopct='%1.0f%%',)
```
